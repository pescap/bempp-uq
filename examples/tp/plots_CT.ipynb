```python
def tensorize(x1, x2=None):
    if x2 is None:
        x2 = x1
    
    x1 = np.array([x1])
    x2 = np.array([x2])
    return np.dot(x1.T, x2)
```

```python

def get_error(L0, L):
    n = 1801
    Sigma_hat_1 = np.zeros([1801, 1801], dtype=np.complex128)
    Sigma_hat_2 = np.zeros([1801,1801], dtype=np.complex128)
    N_hat = 0
    N_hat_sym = 0
    N_max = 0
    #x1_list = []
    #x2_list = []
    #e1_list = []
    #e2_list = []

    for l in range(0, L+1-L0):   
        i0 = L0 + l
        i1 = L - l
        N0 = N_list[i0]
        N1 = N_list[i1]
        u0 = x1_list[i0]
        u1 = x1_list[i1]

        v0 = x2_list[i0]
        v1 = x2_list[i1]

        Sigma_hat_1 += np.dot(np.array([u0]).T, np.array([u1]))
        Sigma_hat_2 += np.dot(np.array([v0]).T, np.array([v1]))
        
        N_hat += N0 * N1
        if i0 <= i1:
            N_hat_sym += N0 * N1
        N_max = np.max([N0 * N1, N_max])


    for l in range(0, L-L0):
        i0 = L0 + l
        i1 = L - l - 1 
        N0 = N_list[i0]
        N1 = N_list[i1]

        u0 = x1_list[i0]
        u1 = x1_list[i1]

        v0 = x2_list[i0]
        v1 = x2_list[i1]

        Sigma_hat_1 -= np.dot(np.array([u0]).T, np.array([u1]))
        Sigma_hat_2 -= np.dot(np.array([v0]).T, np.array([v1]))
        
        N_hat += N0 * N1
        if i0 <= i1:
            N_hat_sym += N0 * N1
        N_max = np.max([N0 * N1, N_max])
        
    Sigma_1 = tensorize(s1, s1)
    Sigma_2 = tensorize(s2, s2)

    notnan_1 = np.where(~np.isnan(Sigma_hat_1))
    notnan_2 = np.where(~np.isnan(Sigma_hat_2))

    Xt1 = Sigma_hat_1[notnan_1]
    St1 = Sigma_1[notnan_1]

    Xt2 = Sigma_hat_2[notnan_2]
    St2 = Sigma_2[notnan_2]

    rel_error_S1 =np.linalg.norm(St2 - Xt1) / np.linalg.norm(St2)
    rel_error_S2 = np.linalg.norm(St1 - Xt2) / np.linalg.norm(St1)
    print(N_list[-1] **2, N_hat, N_hat / N_list[-1] **2 * 100 , 'percents')
    return rel_error_S1, rel_error_S2, N_hat, N_hat_sym, N_max
        
```
