```python
import bempp.api
import numpy as np

import sys
sys.path.append('../../')
from decimal import *

import bemppUQ
from bemppUQ.operators.maxwell import assemble_operators, evaluate_far_field, evaluate_far_field_sd
from bemppUQ.shapes import kite, perturbate
from bempp.api.assembly.blocked_operator import (
    coefficients_of_grid_function_list,
    projections_of_grid_function_list,
    grid_function_list_from_coefficients,
)

from bemppUQ.foa.operators import (
    function_product,
    surface_divergence,
    surface_gradient,
    trace_transformation,
)

bempp.api.global_parameters.assembly.potential_operator_assembly_type = 'dense'
#bempp.api.global_parameters.assembly.boundary_operator_assembly_type = 'dense'
bempp.api.global_parameters.hmat.eps = 1E-4

config = bemppUQ.config.set_case('A')
precision = 20

h = 2.0 * np.pi / (precision * config["k_int"])

print(h)
base_grid = kite(h=h) 
print(base_grid.leaf_view.entity_count(1))

grid_eta, grid_fun = perturbate(base_grid, 0)

# far_field for t=0

transmission_operators = assemble_operators(base_grid, config)
print('FF')
far_field, solution = evaluate_far_field(transmission_operators, config)
print('FFp')
far_field_p, sol_p = evaluate_far_field_sd(base_grid, transmission_operators, config, solution, grid_fun)
```

```python
t = 0.25
config["spaces"] = "maxwell_primal"  # or maxwell
config["osrc"] = True
grid_eta, _ = perturbate(base_grid, t)
transmission_operators_eta = assemble_operators(grid_eta, config)
far_field_eta, solution_eta = evaluate_far_field(transmission_operators_eta, config)
    
    
residual_eta = far_field_eta - far_field
residual_p = far_field_eta - far_field - t * far_field_p
    
err_0 = np.linalg.norm(residual_eta)/np.linalg.norm(far_field_eta)
err_FOA = np.linalg.norm(residual_p)/ np.linalg.norm(far_field_eta)

print(err_0, 'relative norm for the residual eta')
    
```

```python
import matplotlib.pyplot as plt
# Square grid with 10 points per wavelength in each direction

Nx = 2 * int(np.ceil(config['lambda'] * 30))
Ny = int(np.ceil(config['lambda'] * 30))

dim_x = 20
dim_y = 10

# Grid points
xmin, xmax, ymin, ymax = [-dim_x/2.,dim_x/2.,-dim_y/2.,dim_y/2.]
plot_grid = np.mgrid[xmin:xmax:Nx * 1j, ymin:ymax:Ny * 1j]
points = np.vstack((plot_grid[0].ravel(),
                plot_grid[1].ravel(),
                np.zeros(plot_grid[0].size)))
```

```python
k_ext = config['k_ext']
```

```python
electric_pot = bempp.api.operators.potential.maxwell.electric_field(
        transmission_operators[0].domain_spaces[1], points, k_ext
    )
magnetic_pot = bempp.api.operators.potential.maxwell.magnetic_field(
        transmission_operators[0].domain_spaces[0], points, k_ext
    )


electric_pot_eta = bempp.api.operators.potential.maxwell.electric_field(
        transmission_operators_eta[0].domain_spaces[1], points, k_ext
    )
magnetic_pot_eta = bempp.api.operators.potential.maxwell.magnetic_field(
        transmission_operators_eta[0].domain_spaces[0], points, k_ext
    )
```

```python
polarization, direction = config['polarization'], config['direction']

def plane_wave(point):
    return polarization * np.exp(1j * k_ext * np.dot(point, direction))

incident = np.zeros(points.shape)
for i in range(points.shape[1]):
    incident[:, i] = plane_wave(points[:, i])

    
uh = (-electric_pot * solution[1] - magnetic_pot * solution[0]) + incident

uh_eta = (-electric_pot_eta * solution_eta[1] - magnetic_pot_eta * solution_eta[0]) + incident

uh_p= (-electric_pot * sol_p[1] - magnetic_pot_eta * solution_eta[0])
```

```python
uh0 = np.sum(uh, axis=0)
uh0 = uh0.reshape(Nx,Ny)

uh0_eta = np.sum(uh_eta, axis=0)
uh0_eta = uh0_eta.reshape(Nx,Ny)

uh0_p = np.sum(uh_p, axis=0)
uh0_p = uh0_p.reshape(Nx,Ny)
```

```python
import shapely
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize = (12,12))

angle = config['angles']

points_2d = points[:2,:]
ide = (points_2d[0, :]+.75)** 2 + points_2d[1, :] ** 2 <= 3
ide0 = ide.reshape(Nx,Ny)


x1 = np.cos(angle) + (0.65 +t) * np.cos(2 * angle) - 0.65 - t
y1 = (1.5 - 0.25 * t) * np.sin(angle)

trace = np.array([x1,y1])

polygon = Polygon(trace.T)

id_kite_t = np.zeros(points_2d.shape[1])
for i in range(points_2d.shape[1]):
    point = Point(points_2d[0, i], points_2d[1, i])
    id_kite_t[i] = polygon.contains(point)
id_kite_t0 = id_kite_t.reshape(Nx, Ny)

    
ax2.plot(x1, y1, color = 'k', linestyle='-')


x1 = np.cos(angle) + (0.65) * np.cos(2 * angle) - 0.65 
y1 = (1.5 - 0.25) * np.sin(angle)

trace = np.array([x1,y1])

polygon = Polygon(trace.T)

id_kite = np.zeros(points_2d.shape[1])
for i in range(points_2d.shape[1]):
    point = Point(points_2d[0, i], points_2d[1, i])
    id_kite[i] = polygon.contains(point)
id_kite0 = id_kite.reshape(Nx, Ny)

ax1.plot(x1, y1, color = 'k', linestyle='-')
ax3.plot(x1, y1, color = 'k', linestyle='-')


val0 = np.abs(uh0 ** 2)
val_eta0 = np.abs(uh0_eta ** 2)
val_p0 = np.abs(uh0_p ** 2)

vmin = np.min([val0, val_eta0])
vmax = np.max([val0, val_eta0])

vmin=0
vmax=5
matrix = np.fliplr(val0).T
matrix = np.ma.masked_where(id_kite0.T , matrix)
pcm1 = ax1.imshow(matrix,
           extent=[-dim_x/2.,dim_x/2.,-dim_y/2.,dim_y/2.],
           vmin=vmin, vmax=vmax,
           cmap='magma', interpolation='spline16', label=r'$\mathbf{E}$')


matrix = np.fliplr(val_eta0).T
matrix = np.ma.masked_where(id_kite_t0.T , matrix)
pcm2 = ax2.imshow(matrix,
           extent=[-dim_x/2.,dim_x/2.,-dim_y/2.,dim_y/2.],
           vmin=vmin, vmax=vmax,
           cmap='magma', interpolation='spline16', label=r'$\mathbf{E}_t$')


matrix = np.fliplr(val_p0).T
matrix = np.ma.masked_where(id_kite0.T , matrix)
pcm3 = ax3.imshow(matrix,
           extent=[-dim_x/2.,dim_x/2.,-dim_y/2.,dim_y/2.],
           vmin=vmin, vmax=.8,
           cmap='magma', interpolation='spline16', label=r'$\mathbf{E}$')

#ax1.axis('off')
fig.colorbar(pcm1, ax = ax1,pad=0.01)
fig.colorbar(pcm2, ax = ax2,pad=0.01)
fig.colorbar(pcm3, ax = ax3, pad=0.01)

import matplotlib.pyplot as plt
import matplotlib.patches as mpatch


ax1.text(-9.5, 4, r'$\mathbf{E}$', ha='left', bbox={'facecolor': 'white', 'pad': 5}, fontsize=14)
ax2.text(-9.5, 4, r'$\mathbf{E}_t$', ha='left', bbox={'facecolor': 'white', 'pad': 5}, fontsize=14)
ax3.text(-9.5, 4, r"$\mathbf{E}'$", ha='left', bbox={'facecolor': 'white', 'pad': 5}, fontsize=14)

#fig.tight_layout()

#fig.tight_layout()
plt.savefig('plots/PlotsKite.pdf')#, box_inches='tight')
```
