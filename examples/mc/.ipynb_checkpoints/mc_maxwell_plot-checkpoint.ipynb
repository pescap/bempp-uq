```python
from matplotlib import pyplot as plt
import numpy as np



precision = 10
case = 'LF'
t = 0.125

name = (
"results_EM/"
    + "precision"
    + str(precision)
    + 'case'
    + str(case)
    + 't'
    + str(t))

theta = np.linspace(0, 2 * np.pi, 400)
points = np.array([np.cos(theta), np.sin(theta), np.zeros(len(theta))])

def process(M):
    U_mean = np.zeros((3,400), dtype=np.complex128)

    for i in range(M):
        name_t = name + str(i)
        my_dict = np.load(name_t + '.npy', allow_pickle = True, encoding = 'latin1')
        Ut = my_dict.item().get('Ut')
        U_mean += Ut
    
    U_mean /= M
    vart = []

    for i in range(M):
        name_t = name + str(i)
        my_dict = np.load(name_t + '.npy', allow_pickle = True, encoding = 'latin1')
        Ut = my_dict.item().get('Ut')
        
        vart.append( (Ut - U_mean) * np.conj(Ut - U_mean) )


    Uvar = np.sum(np.array(vart), axis=0) / M
    return(U_mean, Uvar, vart)
```

```python
U_mean0, Uvar0, vart0 = process(40)
U_mean1, Uvar1, vart1 = process(100)
```

```python


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16,16), sharex=True, sharey='row')

bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum(np.abs(U_mean0)**2, axis=0))
ax1.plot(theta, bistatic_rcs, label=r'$\hat{U}$', color='r')

bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum( ( np.abs(U_mean0) + np.sqrt(np.real(Uvar0))) **2, axis=0))
ax1.plot(theta, bistatic_rcs, '--' ,label=r'$+\sqrt{\hat{\mathbb{V}}}$', color='b')
bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum( ( np.abs(U_mean0) - np.sqrt(np.real(Uvar0))) **2, axis=0))
ax1.plot(theta, bistatic_rcs, '--' ,label='$-\sqrt{\hat{\mathbb{V}}}$', color='b')
ax1.legend()

bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum( (  np.sqrt(np.real(Uvar0))) **2, axis=0))
ax3.plot(theta, bistatic_rcs, '--' ,label='$\sqrt{\hat{\mathbb{V}}}$', color='b')
ax3.legend()


bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum(np.abs(U_mean1)**2, axis=0))
ax2.plot(theta, bistatic_rcs, label=r'$\hat{U}$', color='r')

bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum( ( np.abs(U_mean1) + np.sqrt(np.real(Uvar1))) **2, axis=0))
ax2.plot(theta, bistatic_rcs, '--' ,label=r'$+\sqrt{\hat{\mathbb{V}}}$', color='b')
bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum( ( np.abs(U_mean1) - np.sqrt(np.real(Uvar1))) **2, axis=0))
ax2.plot(theta, bistatic_rcs, '--' ,label='$-\sqrt{\hat{\mathbb{V}}}$', color='b')
ax2.legend()

bistatic_rcs= 10 * np.log10(4 * np.pi * np.sum( (  np.sqrt(np.real(Uvar1))) **2, axis=0))
ax4.plot(theta, bistatic_rcs, '--' ,label='$\sqrt{\hat{\mathbb{V}}}$', color='b')
ax4.legend()

ax1.grid(color='gray', linestyle='dashed', which='both')
ax2.grid(color='gray', linestyle='dashed', which='both')
ax3.grid(color='gray', linestyle='dashed', which='both')
ax4.grid(color='gray', linestyle='dashed', which='both')
```
